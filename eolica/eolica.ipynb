# Estudo velocidades

```python
data_raw = pd.read_csv('./estacoes/luzilandia.csv', delimiter=';').dropna(subset=['Velocidade do Vento Media']).drop('Unnamed: 11',1)

#data["Data"]= data["Data"].str.split('/', n = 1, expand = True) 

#= pd.DatetimeIndex(data_raw['Data'])
#data_raw['mes'] = data_raw['Data'].dt.month
```

```python
# cada data estará no formato [dia,mes,ano]
data_raw['data'] = [i.split('/') for i in list(data_raw['Data'])]
data_raw['mes'] = [i[1] for i in data_raw['data']]
```

```python
np.mean(data_raw[~data_raw['mes'].isin(meses_chuva)]['Velocidade do Vento Media'])
```

```python
meses_chuva = ['12','1','2','3','4']
meses_seca = ['5', '6', '7', '8', '9', '10', '11']
```

```python
lista_meses = ['Jan','Fev','Mar','Abr','Mai','Jun','Jul','Ago','Set','Out','Nov','Dez']
```

```python
y = data_raw.groupby('mes').mean()['Velocidade do Vento Media'].values
x = ['Jan','Fev','Mar','Abr','Mai','Jun','Jul','Ago','Set','Out','Nov','Dez']

media_seca = np.mean([y[int(i)-1] for i in meses_chuva])
media_chuva = np.mean([y[int(i)-1] for i in meses_seca])


trace1 = go.Bar(x = x, y=y, name = 'Mensal')
trace2 = go.Scatter(x = x, y = 12*[media_seca], name = 'Período de seca', mode = 'lines')
trace3 = go.Scatter(x = x, y = 12*[media_chuva], name = 'Período chuvoso', mode = 'lines')


layout = go.Layout(title = 'Médias mensais da velocidade do vento',
                   yaxis = dict(title = 'Velocidade em m/s'),
                   xaxis = dict(title = 'Meses'))
fig = go.Figure(data = [trace1,trace2,trace3], layout = layout)
pyo.iplot(fig, config=dict(showLink=False))
```

```python
data = []
meses_list = ['Jan','Fev','Mar','Abr', 'Maio','Jun','Jul','Ago','Set','Out','Nov','Dez']

data_raw['mes'] = data_raw['mes'].astype('int')

for mes in np.arange(1,13,1):
    data.append(go.Box(y = data_raw[data_raw['mes'] == mes]['Velocidade do Vento Media'], name = meses_list[mes-1]))

layout = go.Layout(title = 'Boxplot Distribuição velocidades médias', showlegend = False,
                   yaxis = dict(title = 'Velocidade em m/s'),
                   xaxis = dict(title = 'Meses'))

fig = go.Figure(data = data, layout = layout)

pyo.iplot(fig, config=dict(showLink=False))
```
